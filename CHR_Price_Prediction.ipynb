```python
import pandas as pd
import streamlit
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import date
from matplotlib import style
style.use('ggplot')
%matplotlib inline
```

```python
token = ['CHR_USD','ALAPCA-USD','LTO-USD','COTI-USD','CRV-USD','FLUX-USD','RNDR-USD']

df = pdr.DataReader(token, data_source='yahoo', start='2019-01-01', end='2022-03-09')
```

```python
plt.figure(figsize=(16,7))
plt.title('Price History Of CHR Token (Close)', fontsize=15, color='black', fontweight='bold')
plt.plot(df['Close'], color='Red')
plt.ylabel('Price in USD (Close)', fontsize=15, fontweight='bold')
plt.xlabel('Date (Close) ', fontsize=15,fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(16,7))
plt.title('Price History Of CHR Token (Open)', fontweight='bold', fontsize=15)
plt.plot(df['Open'], color='Orange')
plt.xlabel('Date (Open)', fontweight='bold', fontsize=15)
plt.ylabel('Price (Open)', fontweight='bold', fontsize=15)
plt.show()
```
